## Topic Modeling — BERTopic

### Configuration & Tuning

| Version | min_topic_size | Topics | Noise Count | Noise % |
|---------|---------------|--------|-------------|---------|
| v1 | 50 | 79 | 11,173 | 37% |
| v2 | 30 | 127 | 11,773 | 33% |
| Final (reduced) | 50 → reduced | 40 | 11,834 | 39% |

**Decision:** v1 with `reduce_topics(nr_topics=40)` selected — fewer, cleaner topics
with slightly lower noise vs v2. 40 topics is actionable for a support routing system.

**Why pre-computed embeddings?**
Passed `sentence_embeddings.npy` directly to BERTopic — skips redundant embedding
step, saves ~200 seconds per run. BERTopic ran UMAP + HDBSCAN only (58 seconds).

### Topic Quality
- **34 actionable topics** covering: Charging Issues, Wearable Bands, Screen Protector
  Bubbles, Fingerprint Reader, Stylus Issues, Car Vent Mounts, Camera Accessories, etc.
- **5 irrelevant clusters** identified and labeled: Spanish reviews (topic 28),
  meta-commentary (topic 26), generic bad reviews (topic 29, 30), bad cluster (topic 36)
- **3 redundant topic pairs** noted as merge candidates for production:
  - Topics 5, 16, 38 → all fitting-related
  - Topics 9, 10, 19 → all broken/stopped working

### Known Limitations
- **39% noise** — genuinely ambiguous short reviews ("Not working!", "One Star")
  that lack distinctive vocabulary for clustering. These remain available for
  semantic retrieval but are excluded from topic-based routing.
- Some legitimate complaints fell into irrelevant clusters (e.g. row 2:
  *"product is cheaply made, parts missing"* → topic 36/Irrelevant)
- Non-English reviews (Spanish) formed their own cluster — not filtered in preprocessing

## Topic Labelling Fix

**Root cause**: `reduce_topics(nr_topics=40)` was called on topic_model, but the dataframe was populated using `topic_model.topics_` — which reflects the pre-reduction 79-topic assignment. The reduced topic IDs only live on the model object after `reduce_topics()` is called; they are not automatically written back. As a result, `cleaned_reviews.csv` was saved with 78 unique topic IDs instead of 40, and ChromaDB was ingested with those stale labels.

**Detected when**: topic_centroids computed in Phase 6 returned 78 centroids instead of the expected 40.

**Fix**: Re-ran the label assignment using the post-reduction topics, verified `df['topic_id'].nunique() == 40` and `df['topic_label'].isna().sum() == 0`, re-saved CSV, deleted and re-ingested ChromaDB collection.

**Lesson**: After calling `reduce_topics()`, always re-extract topics via `topic_model.topics_` in the same cell, since the attribute is updated in-place. Verify with `nunique()` before saving.

## Second Label Fix — "Stylus/Pen issues" → "Camera lens/protector issues"

**Found during:** Phase 6 LIME explainability (see `explainability.ipynb` and
`chroma_db.ipynb` for full investigation and fix details).

**Issue:** The label `"Stylus/Pen issues"` in `topic_labels` was inherited
from BERTopic's pre-reduction topic name, but the post-`reduce_topics()`
merged cluster actually contains camera lens/protector complaints, not
stylus complaints (verified: 15/15 sampled reviews were camera-protector
related, 0 mentioned a stylus).

**Fix applied here:** Updated `topic_labels` dict entry for this topic_id
from `"Stylus/Pen issues"` to `"Camera lens/protector issues"`, re-ran
`df['topic_label'] = df['topic_id'].map(topic_labels)`, re-saved
`cleaned_reviews.csv`.

**Lesson:** `reduce_topics()` can silently merge semantically unrelated
clusters under a stale name inherited from one of the pre-merge topics.
Manual review of topic labels post-reduction should include spot-checking
a sample of each topic's actual reviews, not just trusting the
auto-generated or manually-assigned name from before reduction.

## Third Label Fix — "Camera/Camera lens issues" → "Generic wear/durability complaints"

**Found during:** Manual audit of topic 17 after fixing topic 14 — checked
for redundancy between the two camera-named topics (14 and 17).

**Investigation:** Sampled 25 reviews from topic 17. None mentioned cameras,
lenses, or photos. Content was generic surface degradation across unrelated
products (scratching, peeling, corrosion) — including one review for a
non-stick cooking pan, indicating minor cross-category data contamination
from the original Phase 1 streaming filter. Checked overlap with screen
protector topics (7, 15) — only 0.66% of topic 17 reviews mention
screen/protector/case, ruling out redundancy with those topics.

**Fix applied:** Relabeled `topic_labels[17]` from
`"Camera/Camera lens issues"` to `"Generic wear/durability complaints"`,
re-applied to df, re-saved CSV.

**Note:** Cross-category contamination (the pan review) was not removed —
documented as a known data limitation rather than re-filtered, since it
affects ~1 row out of 30,157.
